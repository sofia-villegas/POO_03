10.- ¿Qué películas fueron dirigidas por Brenda Chapman?

```python
# Ejecuta una consulta SQL
# Solo pide la infofrmación donde el director de la película es 'Brenda Chapman'
cur.execute("SELECT original_title FROM movies JOIN directors ON directors.id = movies.director_id WHERE directors.name = 'Brenda Chapman';")

# Recupera los resultados
directed_by = cur.fetchall()

# Imprime los títulos de las películas dirigidas por Brenda Chapman
print("Brenda Chapman \n\n \t", directed_by)
```

11.- ¿Qué director ha realizado más películas?

```python
# Ejecuta una consulta
#solo pide la info donde se cuenta el número de películas dirigidas por cada director
# ademas agrupa por director_id y se ordena en orden descendente según el recuento del nombre del director
cur.execute("SELECT name FROM directors JOIN movies ON directors.id = movies.director_id GROUP BY director_id ORDER BY count(name) DESC limit 1;")

# Recupera el nombre del director
directed_by = cur.fetchall()

# Imprime el nombre del director con más películas
print("El director con más películas \n\n \t", directed_by)
```

12.- Enumere las 10 películas más populares

```python
# Ejecuta una consulta SQL
# solo pide info de popularidad y ingresos de las películas de la tabla 'movies' unidas con la tabla 'directors'
# Los resultados se ordenan por popularidad de forma descendente.
cur.execute('SELECT original_title, name, release_date, popularity, revenue FROM movies JOIN directors ON movies.director_id = directors.id ORDER BY popularity DESC')

# Recupera los resultados
mostPopular = cur.fetchall()

# Convierte los resultados en un DataFrame
mostPopular = pd.DataFrame(mostPopular, columns=['original_title', 'director_name', 'release_date', 'popularity', 'revenue'])

# Muestra las primeras diez filas del DataFrame
mostPopular.head(10)
```

15.- Enumere los nombres de los directores con la cantidad de películas y los ingresos.

```python
# Selecciona el nombre del director, cuenta de títulos originales y suma de ingresos de películas unidas por director y ordenadas por ingresos descendentes.
cur.execute('SELECT name, COUNT(original_title), SUM(revenue) FROM directors JOIN movies ON directors.id = movies.director_id GROUP BY name ORDER BY SUM(revenue) DESC')

# Recupera los resultados
mostProfitableDirector = cur.fetchall()

# Crea un DataFrame
mostProfitableDirector = pd.DataFrame(mostProfitableDirector, columns=['director_name', 'movies', 'revenue'])

# Muestra las primeras diez filas del DataFrame
mostProfitableDirector.head(10)
```
